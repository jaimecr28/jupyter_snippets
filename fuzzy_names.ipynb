```python
#Bibliotecas necessarias
import pandas as pd
import numpy as np
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
data_rh = pd.read_csv("loja1.csv" ,";")
```

```python
nomes_rh = data_rh['Nome']
```

```python
data_resp = pd.read_csv("loja1_respostas.csv", ";")
```

```python
nomes_respostas = data_resp['NOME']
```

```python
match_names=[]
nomes_respostas = data_resp['NOME']
nomes_rh = data_rh['Nome']
for nome in nomes_respostas:
    print(nome)
    nome_rh = process.extractOne(nome, nomes_rh)
    print(nome_rh)
    match_names.append([nome, nome_rh[0]])
```

```python
match_data = pd.DataFrame(match_names, columns=['Resposta', 'RH'])
```

```python
match_data.to_excel('match_loja1.xls', index=False, encoding='utf-8')
```
